### 출력문 
파이썬에서 출력문은 'print()' 명령어를 사용합니다

```python
# '#' 이것은 주석입니다.
```

```python
# 파이썬 명령어는 마지막에 세미콜론 붙이지 않음. ';' (X)
print(10)
print('문자는 홀따옴표로 감싼다.')
print("문자는 쌍따옴표도 가능하다.")
```

```python
# < 연산 결과의 출력 > #
print(10 + 5) # [15]
print(10 / 4) # [2.5] 연산 결과 : 실수 => '실수 값' 나옴
print(10 // 4) # [2] '//' : (1) 나눗셈의 '몫' (계산)
print(10 % 4) # [2] '%' : (2) 나눗셈의 '나머지' (계산)
print('java' + 'python') # [javapython] (3) '문자' + '문자' => '문자열' 나옴
# print('python' + 123) # 파이썬은 '문자' + '숫자' => 안 된다. (X)
```

```python
# < print()의 활용 > #
print(10, 'hello', 10.5)

# 'end' 이것은 '개행 없이 (옆) 에다 쓰기' 입니다.
# 즉 (옆) or (마침표.) or (아래\n)
# print() 함수는 'end' 매개변수로 출력 후, 실행코드를 작성 할 수 있다.
print('hi', end='\n') 
print('hi', end='')
print('hi', end='.')
```

```python
# < 연습 - 출력해보기 > #
myName = '홍길동'
myAge = 20

#이름은 홍길동이고, 나이는 20살입니다

# 포맷함수 [ f'' ] = '백틱``과 같은 역할' #

print(f'이름은 {myName}이고, 나이는 {myAge}살 입니다')
```
