조건문

```python
# <조건문> - (if절) #

# (1) 파이썬은 중괄호도 없다.=> 
# ':'으로 한다. 
# '=' (X)
 
num1 = 3

# (2)'들여쓰기': 파이썬의 if문 실행영역을 판단한다.
if num1 > 1 :
  print(1)
  print(5)

print(10)
```

```python
# <조건문> - (if절 / elif절 / else절) #

num2 = 5

if num2 > 1 :
  print(1)
elif num2 > 3 :
  print(2)
else :
  print(3)
```

```python
# < 연습 1 - 문제 > #

# ------------------------------------------------------
#                   (홀/짝) 판별기
# ------------------------------------------------------
# 키보드로 정수를 하나 입력받아
# 입력 받은 수가 짝수면, '짝수'를 출력
# 입력 받은 수가 홀수면, '홀수'를 출력
# ------------------------------------------------------
# 키보드 입력을 위해 input()을 사용함 
# (But 키보드로 입력받은 모든 데이터는 무조건 '문자'로 인식)
# 문자 => 정수로 변환 : int('5') => 5
# ------------------------------------------------------

num3 = int(input('정수 1개를 입력해주세요 : '))
if num3 % 2 == 0 : 
  print('짝수')
else :
  print('홀수')
```
