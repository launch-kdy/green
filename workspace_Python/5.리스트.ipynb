```python
# < List > #

list1 = [] # 빈 리스트 생성
list2 = [1, 2, '안녕', 5.5] # 모든 자료형 가능
list3 = [1, 2, [3, 4, 5]] # 배열 안에 배열 가능
```

```python
# < 연습 1 > #

print(list2[1]) # list2 = [1, (2), '안녕', 5.5] # (2를 출력)
print(list3)
```

```python
# < List > #

# < List > 에 데이터 (넣기)


list1 = []
list1.append(1) # '.append()함수로 넣기' ex) 1 을 넣어보자.
print(list1[0])
list1.append(2)
list1.append(3)
print(list1)
# 1) '.append()' 함수 : 리스트의 마지막 요소에 데이터가 추가된다.

list1.insert(1, 4)
print(list1)
# 2) 'insert(넣을 index 칸 위치 - 정수로 적기, 넣고싶은 데이터) : 원하는 위치에 데이터가 추가된다.
```

```python
# < List > #

# < List > 에 데이터 (삭제)

list1 = [1,2,3,4,5]

del list1[2] # ex) list1 변수의 2번 요소(index 칸)의 데이터를 삭제
print(list1)

# 주의사항
del list1 # 'list1' 이라는 변수 자체를 제거가 지워지기 때문에 출력시 오류가 발생함
```

```python
# < List > #

# < List > 에 데이터 (수정)

list1 = [1,2,3]

# ex) 2 를 22222 으로 바꾸자.
list1[1] = 2222
print(list1)
```

```python
# < List > #

# '슬라이싱' : 리스트에서 특정 영역만 (추출)하는 문법

# [추출시작 index : 추출 마지막 index는 (x)

list2 = [1,2,3,4,5]
print(list2[1:4]) # 1번 칸 index 에서 4번 index(전)까지 !
print(list2[:3]) # 시작 index (칸)을 안 넣으면 => 자동으로 '0번째 칸'으로 맞춰줌
print(list2[2:])
print(list2[:]) # 모든데이터를 (출력)
print(list2[2:-1])
```
