```python
# < 반복문 > #

# 파이썬은 '일반 for문' 없음 (X)

# 파이썬은 'for-each문' 밖에 없음 (O)
```

```python
# < 반복문 > #

for e in [1,2,3] :
  print(e)
  print(1)
print(1000)
```

```python
# < 반복문 > for in

list1 = [1,2,3]
for e in list1 :
  print(e)
```

```python
# < 반복문 > for range ( [ X ] < [ ? ] < [ Y ] )

# 0부터 10보다 작은 동안 반복하겠다

# range 는 특정 데이터로 반복한다는 의미보다는,
# 단순히 똑같은 작업을 몇번 반복 할 때 사용

for e in range(0, 10) : # == for(int i = 0 ; i < 10 ; i++)
  print(e)
```

```python
# < 반복문 > #

# 파이썬은 일반 for문이 없기 때문에 반복 횟수를 알 수 있는 index 데이터가 없음.

# 그래서 for-each문에 'index를 알 수 있는 문법'이 추가되어 있다.

list = [1,2,3,4,5]
for i, e in enumerate(list1) :
  print(f'i = {i}, e = {e}')
```
