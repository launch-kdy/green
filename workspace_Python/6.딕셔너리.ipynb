```python
# < (딕셔너리 = 객체) dictionary >

# 여러 데이터를 저장 할 수 있고, key와 value 의 쌍으로 데이터가 저장된다.

# 걍 객체다.
```

```python
# (1) 딕셔너리 데이터 (생성)

dic1 = {} # 빈 딕셔너리 데이터 생성
dic2 = {
  'stuNum' : 222,
  'name' : '김자바',
  'age' : 20
}
```

```python
# (2) 딕셔너리 데이터 (읽기)

# 전체 읽기
print(dic2)

# 뽑아 읽기
print(dic2['name'])
```

```python
# (3) 딕셔너리 데이터 (추가 / 변경 / 삭제)

# 예시 ------------
phone = {
  'model' : 's25',
  'price' : 10000
}
# -----------------

# 데이터 추가 ( 추가 ) [없는'하고싶은 key'를 걍 적으면, 자동으로 추가됨]
phone['color'] = '화이트'
print(phone)

# 데이터 변경 ( 변경 ) [기존 '존재하는 key'에 데이터를 넣으면 자동으로 변경됨]
phone['price'] = '10100'
print(phone)

# 데이터 삭제 ( 삭제 del ) [기존 '존재하는 key'앞에 del 를 적으면 자동으로 변경됨]
del phone['model']
print(phone)

```

```python
# (3) 딕셔너리의 모든 데이터에 접근하는 방법

# 예시 ------------
monitor = {
  'brand' : 'samsung',
  'price' : 10000,
  'color' : '화이트'
}
# -----------------


# 딕셔너리 데이터는 for문을 이용하여 모든 데이터에 접근 할 수 있음
# 딕셔너리 데이터를 for문으로 조회하면 only 'key값만' 나옴

# 딕셔너리 데이터를 for문으로 조회하면 only 'key값만' 나오는데, [value도 조회하고 싶을 때]
for e in monitor :
  print(f'key = [e], value = {monitor[e]}')

```

```python
# < 문제 >

# 예시 -----------------------------------------
myInfo = {
  '이름' : '크롱',
  '나이' : 5,
  '취미' : ['과자먹기', '블럭쌓기', '쥬스마시기'],
  '가족' : [
    {
      '관계' : '친구1',
      '이름' : '뽀로로'
    },
    {
      '관계' : '친구2',
      '이름' : '루피'
    }
  ]
}
# ----------------------------------------------

# 문제 1 : myInfo 딕셔너리에서 '모든 취미' 출력
print(myInfo['취미'])

# 문제 2 : myInfo 딕셔너리에서 '취미의 갯수' 출력
# 힌트) : 배열의 길이(크기)를 알고 싶을 때는 len() 함수 사용
# ex) len([1,2,3]) => 3 
print(len(myInfo['취미']))


# 문제 3 : myInfo 딕셔너리에서 '마지막 취미'를 출력
print(myInfo['취미'][-1])


# 문제 4 : myInfo 딕셔너리에서 '친구1의 이름'을 출력
print(myInfo['가족'][0]['이름'])


# 문제 5 : myInfo 딕셔너리에서 '친구2의 나이'를 추가 (나이는 5)
myInfo['가족'][1]['나이'] = 5
print(myInfo)

```
