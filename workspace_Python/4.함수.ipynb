```python
# < 함수 > (선언)

def hamsuhamsu():
  print(1)
  print(2)
```

```python
# < 함수 > (호출)
hamsuhamsu()
```

```python
# < 함수 > (리턴)
def reeturn(a, b):
  print(a + b)
  return a + b
```

```python
# < 함수 > (리턴 - 출력)
result = reeturn(10, 20)
print(result)
```

```python
# 파이썬의 < 함수 > 는 (리턴 데이터 - 여러개도 가능)
def getNumber():
  return 10, 20, 30
```

```python
r1, r2, r3 = getNumber()
print(r1, r2, r3)
```

```python
# 파이썬의 < 함수 > 는 (매개변수에 데이터를 전달 할 때 - 매개변수 이름 활용 가능 '순서대로 넣기 싫을 때')
def printInfo(name, age):
  print(f'이름은 {name}이고, 나이는 {age}입니다')
```

```python
printInfo('포비', 5)

printInfo(name = '포비', age = 5)

printInfo(age = 3, name='에디')
```
